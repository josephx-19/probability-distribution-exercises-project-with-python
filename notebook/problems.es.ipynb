# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import norm, uniform, binom, randint


# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()

print(data)
```

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
#TODO
from scipy.stats import norm, uniform, binom
import numpy as np

# Función para la PDF de una distribución uniforme (distribución continua).
def duniforme(x, size, prob = 0.5):
    """
    Estima la probabilidad para una distribución uniforme continua.
    """
    result = uniform.pdf(x, loc=0, scale=1)  # Probabilidad para una distribución uniforme continua
    return result

def dnorm(x, size, prob = 0.5):
    """
    Estima la probabilidad para una distribución uniforme continua.
    """
    result = norm.pdf(x, loc=0, scale=1)  # Probabilidad para una distribución normal 
    return result

def pnormal(q, size, prob = 0.5):
    """
    Calcula la función de distribución acumulada (CDF) para la distribución normal.
    """
    result = norm.cdf(q, loc=0, scale=1)  # CDF para la distribución normal
    return result

def puniforme(q, size, prob = 0.5):
    """
    Calcula la función de distribución acumulada (CDF) para la distribución uniforme.
    """
    result = uniform.cdf(q, loc=0, scale=1)  # CDF para la distribución uniforme
    return result

def qnormal(p, size, prob = 0.5):
    """
    Calcula el cuantil (percentil) para la distribución normal.
    """
    result = norm.ppf(p, loc=0, scale=1)  # Cuantil para la distribución normal
    return result

def quniform(p, size, prob = 0.5):
    """
    Calcula el cuantil (percentil) para la distribución uniforme.
    """
    result = uniform.ppf(p, loc=0, scale=1)  # Cuantil para la distribución uniforme
    return result

def rnormal(size):
    """
    Genera valores aleatorios a partir de una distribución normal.
    """
    result = norm.rvs(loc=0, scale=1, size=size, random_state=None)  # Valores aleatorios de la normal
    return result

def runiform(size):
    """
    Genera valores aleatorios a partir de una distribución uniforme.
    """
    result = uniform.rvs(loc=0, scale=1, size=size, random_state=None)  # Valores aleatorios de la uniforme
    return result
```

```python
size = 10
prob = 0.2
data_norm = rnormal(2000)
data_uniform = runiform(2000)

pdf_uni = duniforme(data_uniform, size=size, prob=prob)  
cdf_uni = puniforme(data_uniform, size=size, prob=prob)
ppf_uni = quniform([0.1, 0.5, 0.9], size=size, prob=prob)

# CDF y PPF para la distribución normal
cdf_norm = pnormal(data_norm, size=size, prob=prob)
ppf_norm = qnormal([0.1, 0.5, 0.9], size=size, prob=prob)
pdf_norm = dnorm(data_norm, size=size, prob=prob)

# Imprimir los resultados
np.random.seed(42)

print("PDF Uniforme:", pdf_uni)
print("CDF Uniforme:", cdf_uni)
print("PPF Uniforme:", ppf_uni)
print()
print("CDF Normal:", cdf_norm)
print("PPF Normal:", ppf_norm)
print("PDF Normal:", pdf_norm)

```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Gráfico de la PDF Uniforme
axes[0].plot(data_uniform, pdf_uni, label="PDF Uniforme", color='blue')
axes[0].set_title("PDF - Distribución Uniforme")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Densidad de Probabilidad")
axes[0].legend()

# Gráfico de la CDF Uniforme
axes[1].plot(data_uniform, cdf_uni, label="CDF Uniforme", color='green')
axes[1].set_title("CDF - Distribución Uniforme")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Función de Distribución Acumulada")
axes[1].legend()

# Gráfico de PPF Uniforme (percentiles)
axes[2].scatter([0.1, 0.5, 0.9], ppf_uni, label="PPF Uniforme", color='red')
axes[2].set_title("PPF - Distribución Uniforme")
axes[2].set_xlabel("Probabilidad")
axes[2].set_ylabel("Cuantil")
axes[2].legend()


plt.show()

fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Gráfico de la PDF Normal
axes[0].plot(data_norm, pdf_norm, label="PDF Normal", color='purple')
axes[0].set_title("PDF - Distribución Normal")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Densidad de Probabilidad")
axes[0].legend()

# Gráfico de la CDF Normal
axes[1].plot(data_norm, cdf_norm, label="CDF Normal", color='orange')
axes[1].set_title("CDF - Distribución Normal")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Función de Distribución Acumulada")
axes[1].legend()

# Gráfico de PPF Normal (percentiles)
axes[2].scatter([0.1, 0.5, 0.9], ppf_norm, label="PPF Normal", color='blue')
axes[2].set_title("PPF - Distribución Normal")
axes[2].set_xlabel("Probabilidad")
axes[2].set_ylabel("Cuantil")
axes[2].legend()


plt.show()
```
